### QUESTION 1:

### Ideas that cannot be tested statistically are ideas that do not initially provide us with something quantifiable or even testable in terms of a population parameter. For example, saying "Cars in new york are expensive" isn't a hypothesis, since were not providing any information that makes it testable. Lets change that to "Cars in new york cost $50000" Now we have something quantifiable and testable since were given a number in this situation. So the key criteria would be to make sure that the hypothesis is 1. testable, and 2. quantifiable. 

### Defining a good null hypothesis is when we effectively define a specific testable population parameter. $50000 in the example above is a specific testable population parameter. A null hypothesis should also be falsifiable, and represent a no effect situation where we can compare the data we observed.

### The difference between a null hypothesis and an alternative hypothesis is that a null hypothesis is simply what we have assumed to be true until proven false, the alternate hypothesis is everything else. So if the car doesnt cost $50000, the null hypothesis is false and the alternate hypothesis is that it costs something else.

### We imagine that we are in a world where the null hypothesis is true so that we have a baseline for what we are expecting to happen, and what we are trying to prove is actually false. When we calculate the p-value, we are seeing the probability of observing our null hypothesis in a world where we assume its true. We then use the calculated p-value to determine if we have enough evidence to reject the null hypothesis or to accept it.

### QUESTION 4:

### A smaller p-value makes the null hypothesis seem more ridiculous because it indicates that our observed statistic (from the sample) is increasingly unlikely when we assume that the null hypothesis is true. The null hypothesis describes a scenario of no effect or of no difference. When we calculate the p-value, we observe how extreme our results are if we are to assume that the null hypothesis is true. If our p-value is very small, this is indicating that the world in which we assume the null hypothesis is true, is unlikely based on our sample statistics (It should be known that a p-value doesnt indicate certainty, it will never be certain.) What the p-value does provide is that it becomes increasingly improbable to assume that the null hypothesis is true.

### Fishers Tea Experimentation

### Problem Introduction:

### In this problem we are replicaiting the famous expirment where we test the ability of people to tell whether or not milk had been poured into the cup before or after the tea. In this case we are using STA130 students.

### Relationship between this experiment and Fishers:

### In fishers experiment, Bristol was, with complete accuracy, able to identify which of the cups of tea had the milked poured in before the tea. In our expirement, we see that only 49/80 students could do this, so this begs the question, "Is it a coincidence, or is it possible?"

### The Null and Alternative Hypothesis:

### The Null Hypothesis: The ability of people to guess the order of milk and tea being poured into the cup has a 50% accuracy.
### Informally: Students will not be able to guess how their tea was made.

### The Alternative Hypothesis: The ability of people to guess the order of milk and tea being poured into the cup is greater than 50% accuracy.

### Our observed statistic:

### Out of 80, there were 49 correct guesses, which tells us that we have a higher than 50% accuracy. But, this isnt enough information, what we should be doing is taking this data and sampling from it to get more information. Lets try and find the p-value to reject or confirm the null hypothesis.





```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
n_students = 80
n_correct = 49

# Observed proportion of correct responses
observed_proportion = n_correct / n_students

# Bootstrap parameters
n_bootstrap_samples = 10000
bootstrap_proportions = []

# Bootstrapping process
for _ in range(n_bootstrap_samples):
    # Create a bootstrap sample
    bootstrap_sample = np.random.choice([1, 0], size=n_students, replace=True, p=[observed_proportion, 1-observed_proportion])
    
    # Calculate the proportion of correct responses in the bootstrap sample
    bootstrap_proportion = np.mean(bootstrap_sample)
    bootstrap_proportions.append(bootstrap_proportion)

# Calculate the p-value
p_value = np.mean(np.array(bootstrap_proportions) >= observed_proportion)

# Print results
print("Observed Proportion:", observed_proportion)
print("Bootstrap P-value:", p_value)

# Optional: Visualize the bootstrap distribution
plt.hist(bootstrap_proportions, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(observed_proportion, color='red', linestyle='dashed', linewidth=2, label='Observed Proportion')
plt.title('Bootstrap Distribution of Proportions')
plt.xlabel('Proportion of Correct Responses')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### What is happening in this code is we are bootstrapping. Creating many samples with replacements from our original sample. With all these other samples (which are the same size as the original) we calculate the statistic we are looking for, which is the accuracy of students guessing how their tea was made.

### The observed proportion is 0.61, meaning the students had an accuracy of 61%. That is higher than 50! But the important thing to look at is the p-value, which is 0.5434. This is not enough evidence to reject the null hypothesis.

### Conclusion:

### We fail to reject the null hypothesis. There is not enough strong evidence to claim that the null hypothesis can be rejected and/or is unlikely. This means that we continue believing the null hypothesis and that we are unable to tell how our tea was made. This means that we can conclude the 49/80 accuracy was (most likely, not certain) a coincidence.
